```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("SPY.csv")
df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
del df['Date']
del df['Close']
cormap = df.corr()
fig, ax = plt.subplots(figsize=(5,5))
sns.heatmap(cormap, annot = True)
```

```python
sns.pairplot(df)
plt.tight_layout()
```

```python
X = df.drop(['Adj Close'], axis=1)
y = df['Adj Close']
X
```

```python
from sklearn.preprocessing import MinMaxScaler


# This is to standardize the data since it varies in range
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression

# model training

model_1 = LinearRegression()
model_1.fit(X_train, y_train)
```

```python
y_pred_1 = model_1.predict(X_test)
pred_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_1})
pred_df.head()
```

```python
from sklearn.metrics import r2_score, accuracy_score, explained_variance_score, max_error
Acc = []  # Accuracy
print("Accuracy score of the predictions: {0}".format(r2_score(y_test, y_pred_1)))
Acc.append(r2_score(y_test, y_pred_1))

print("Explained Variance score is: {0}".format(explained_variance_score(y_test, y_pred_1)))
print("Max Error is: {0}".format(max_error(y_test, y_pred_1)))
```

```python
plt.figure(figsize=(8,8))
plt.ylabel('Close Price', fontsize=16)
plt.plot(pred_df)
plt.legend(['Actual Value', 'Predictions'])
plt.show()
```

## CNN

```python
from tensorflow.keras import Sequential,utils
from tensorflow.keras.layers import Flatten, Dense, Conv1D, MaxPool1D, Dropout
```

```python
X_train = np.array(X_train).reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = np.array(X_test).reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
def CNN():    
    model = Sequential()
    
    model.add(Conv1D(32, kernel_size=(3,), padding='same', activation='relu', input_shape = (X_train.shape[1],1)))
    model.add(Conv1D(64, kernel_size=(3,), padding='same', activation='relu'))
    model.add(Conv1D(128, kernel_size=(5,), padding='same', activation='relu'))
    model.add(Flatten())
    
    # After some trial and error, two dense layers "Relu" activation functions worked best at minimizing loss
    model.add(Dense(50, activation='relu'))
    model.add(Dense(20, activation='relu'))
    model.add(Dense(units = 1))
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

```python
model_2 = CNN()
model_2.fit(X_train, y_train, epochs=50, validation_split=0.3)
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
pred_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_2.flatten()})
pred_df.head()
```

```python
print("Accuracy score of the predictions: {0}".format(r2_score(y_test, y_pred_2)))
Acc.append(r2_score(y_test, y_pred_2))

print("Explained Variance score is: {0}".format(explained_variance_score(y_test, y_pred_2)))
print("Max Error is: {0}".format(max_error(y_test, y_pred_2)))
```

```python
plt.figure(figsize=(8,8))
plt.ylabel('Close Price', fontsize=16)
plt.plot(pred_df)
plt.legend(['Actual Value', 'Predictions'])
plt.show()
```

## ANN

```python
# Artificial Neural Network model char, since Relu worked best before, I reapplied this
def ANN(inp_dim):
    
    model = Sequential()
    
    model.add(Dense(20, input_dim=inp_dim, kernel_initializer='normal', activation='relu'))
    model.add(Dense(25, kernel_initializer='normal', activation='relu'))
    model.add(Dense(10, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

```python
model_3 = ANN(inp_dim=4)
model_3.fit(X_train, y_train, epochs=50, validation_split=0.2)
```

```python
y_pred_3 = model_3.predict(X_test)
```

```python
pred_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_3.flatten()})
pred_df.head()
```

```python
print("Accuracy score of the predictions: {0}".format(r2_score(y_test, y_pred_3)))
Acc.append(r2_score(y_test, y_pred_3))

print("Explained Variance score is: {0}".format(explained_variance_score(y_test, y_pred_3)))
print("Max Error is: {0}".format(max_error(y_test, y_pred_3)))
```

# Time Series to improve Accuracy

```python
close = df.reset_index()['Adj Close']
close.head()
```

```python
# Checking to see the shape of adjusted stock price close
plt.plot(close)
plt.show()
```

```python
time_step = 30
X = []
y = []

for i in range(len(close)-time_step-1):
    X.append(close[i:(i+time_step)])  # in month format
    y.append(close[(i+time_step)])    # single value

X = np.array(X)
y = np.array(y)

```
